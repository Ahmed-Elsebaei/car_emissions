```python
# The query above has shown that all nulls are in EV cars
# After research, Almost all EVs have a single speed transmission. Also the term "single speed automatic" is common to describe  this mechanism
# So, we can change all nulls in transmission column with "1A"
```

```python
df.transmission.fillna('1A', inplace = True)
df.info()
```
